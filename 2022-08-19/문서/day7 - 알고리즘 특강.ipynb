# 원형 큐
<br>

큐의 처음과 끝이 연결된 구조

데이터의 양이 늘어나면 오버헤드가 발생하니까... 이거 쓰면 오버헤드 X

일반 큐의 초기값은 -1, 원형 큐는 0

````python
# 함수
def isQueueFull():
    global SIZE, queue, front, rear
    if ( (rear+1) % SIZE == front) :                     # rear와 front가 같다고 하면 큐가 꽉 찼다고 인식 x
        return True                                      # %n 을 해서 rear의 값이 n엥서 다시 0으로 감
    else:
        return False
    
def enQueue(data):
    global SIZE, queue, front, rear
    if (isQueueFull()):
        print("큐 꽉참")
        return
    rear = (rear + 1) % SIZE
    queue[rear] = data

def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear):
        return True
    else:
        return False
    
def deQueue():
    global SIZE, queue, front, rear
    if (isQueueEmpty()):
        print('큐 비었음')
        return None
    front = ( front + 1 ) % SIZE
    data = queue[front]
    queue[front] = None
    return data
    
def peek():
    global SIZE, queue, front, rear
    if (isQueueEmpty()):
        print('큐 비었음')
        return None
    return queue[(front+1) % SIZE]


# 전역변수
SIZE = 5
queue = [None for _ in range(SIZE)]
front = 0
rear = 0


# 메인코드
enQueue('화사')
enQueue('솔라')
enQueue('문별')
enQueue('휘인')
#enQueue('선미')
print('출구 <--', queue, '<-- 입구')

print('식사 손님 :',deQueue())
print('식사 손님 :',deQueue())
print('출구 <--', queue, '<-- 입구')

enQueue('재남')
print('출구 <--', queue, '<-- 입구')

enQueue('제니')
print('출구 <--', queue, '<-- 입구')

enQueue('영우')
print('출구 <--', queue, '<-- 입구')
# print('식사 손님:',deQueue())
# print('식사 손님:',deQueue())
# print('**준비하세요**',peek())
# print('식사 손님:',deQueue())
# print('식사 손님:',deQueue())
# print('식사 손님:',deQueue())

# print('출구 <--', queue, '<-- 입구')
# print('식사 손님:',deQueue())
````

# 이진 트리

트리 구조란?

### 이진 트리 
<br>

모든 노드의 자식이 2개 이하로 구성된 트리

다른 트리는 너무 표현하기 힘들어서 간단한 이진 트리 씀

서브 트리 : 부모 노드의 Left, Right 자식 트리 덩어리들 

포화 이진 트리, 완전 이진 트리, 편향 이진 트리....

이진 트리의 노드 구조 : 왼쪽 링크 + 데이터 + 오른쪽 링크

````python
# 높이가 2고 데이터가 6인 완전 이진 트리 생성 

# 함수
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
        

# 전역변수


# 메인코드
node1 = TreeNode()
node1.data = '화사'

node2 = TreeNode()
node2.data = '솔라'
node1.left = node2

node3 = TreeNode()
node3.data = '문별'
node1.right = node3

node4 = TreeNode()
node4.data = '휘인'
node2.left = node4

node5 = TreeNode()
node5.data = '쯔위'
node2.right = node5

node6= TreeNode()
node6.data = '선미'
node3.left = node6

print(node1.data)
print(node1.left.data, node1.right.data)
print(node1.left.left.data, node1.left.right.data, node1.right.left.data)
````

# 그래프
<br>

버스 정류장과 노선이 함께 포함된 형태, 링크드인(Linked in)과 같은 사회관계망 서비스의 연결 등의 형태

---

무방향 그래프 : 간선에 방향성이 없는 그래프
> G1 그래프, G2 그래프....

그래프에서는 노드가 아닌 **정점(voltex)**, 엣지가 아닌 간선

G1, G2의 정점 집합 표현
```python
V(G1) = {A,B,C,D}
V(G2) = {A,B,C,D}
```

G1, G2의 간선 집합 표현
```python
V(G1) = {(A,B),(A,C),...(C,D)}
V(G2) = {(A,B),C,D}
```
---

방향 그래프

---

가중치 그래프 : 간선마다 가중치가 다르게 부여된 그래프

### 깊이 우선 탐색
<br>

▪ 그래프의 모든 정점을 한 번씩 방문하는 것을 그래프 순회(Graph Traversal)라고 함

▪ 그래프 순회 방식은 깊이 우선 탐색, 너비 우선 탐색이 대표적

> 깊이 우선 방식은 더 이상 갈 데 없으면 왔던 길로 돌아감

---

### 그래프의 인접 행렬 표현
<br>

▪ 그래프를 코드로 구현할 때는 인접 행렬을 사용

▪ 인접 행렬은 정방형으로 구성된 행렬로 정점이 4개인 그래프는 4×4로 표현

▪ 무방향 그래프의 인접 행렬은 대각선을 기준으로 서로 대칭

````python
# 함수
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [ [ 0 for _ in range(size)] for _ in range(size)]  # 2차원 빈 배열
        



# 전역변수
G1 = None



# 메인코드
A = 0; B = 1; C = 2; D =3;
G1 = Graph(4)
G1.graph[A][B] = 1                # A와 B간의 간선
G1.graph[A][C] = 1
G1.graph[A][D] = 1

G1.graph[B][A] = 1
G1.graph[B][C] = 1

G1.graph[C][A] = 1
G1.graph[C][B] = 1
G1.graph[C][D] = 1

G1.graph[D][A] = 1
G1.graph[D][C] = 1

for row in range(4):
    for col in range(4):
        print(G1.graph[row][col], end = ' ')
    print()
````

### 선택 정렬
<br>

여러 데이터 중에서 가장 작은 값을 뽑는 작동을 반복하여 값을 정렬

---

최솟값 찾는 방법

1. 배열의 첫 번째 값을 가장 작은 값으로 지정
2. 그 값을 다음 값이랑 비교하면서 변경하거나 그대로 둠
3. 마지막 값까지 싹 다 비교하고 현재 가장 작은 값이 가장 작은 값

````python
# 최솟값 찾기

# 함수
def findMinIndex(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if (ary[minIdx] > ary[i]):
            minIdx = i
    return minIdx
            
# 전역변수
testAry = [55, 88, 33, 77]

# 메인코드
minPos = findMinIndex(testAry)
print('최솟값 -->', testAry[minPos])
````

````python
# 완전판

# 함수
import random

def findMinIndex(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if (ary[minIdx] > ary[i]):
            minIdx = i
    return minIdx


# 전역변수
before = [random.randint(1,200) for _ in range(20)]
after = []


# 메인코드
print('정렬전 -->', before)
for _ in range(len(before)):
    minPos = findMinIndex(before)
    after.append(before[minPos])
    del(before[minPos])                       # 이걸 해줘야 before에서 빠지고 after로 들어가는 꼴
    
print('정렬후 -->', after)
````

````python
# 개선된 선택 정렬 구현

# 함수
import random

def selectionSort(ary):
    n = len(ary)
    for i in range(0, n-1):        # Cycle (큰 회전)
        minIdx = i
        for k in range(i+1, n):
            if (ary[minIdx] > ary[k]):
                minIdx = k     
        ary[i], ary[minIdx] = ary[minIdx], ary[i]
    
    return ary
    
# 전역변수
dataAry = [random.randint(1,200) for _ in range(20)]


# 메인코드
print('정렬전 -->', dataAry)
dataAry = selectionSort(dataAry)
print('정렬후 -->', dataAry)
````

# 알고리즘 - 정렬

어떤 집합에서 원하는 것을 찾는 것으로, 탐색이라고도 함

▪ 검색에는 순차 검색, 이진 검색, 트리 검색 등이 있음

---

### 검색 알고리즘의 종류
<br>

순차 검색

▪ 검색할 집합이 정렬되어 있지 않은 상태일 때

▪ 처음부터 차례대로 찾아보는 것으로, 쉽지만 비효율적임

▪ 집합의 데이터가 정렬되어 있지 않다면 이 검색 외에 특별한 방법 없음

이진 검색

▪ 데이터가 정렬되어 있다면 이진 검색도 사용 가능

▪ 순차 검색에 비해 월등히 효율적이라 데이터가 몇 천만 개 이상이어도 빠르게 찾아낼 수 있음

트리 검색(얘는 했던거라 빼도 댐)

▪ 데이터 검색에는 상당히 효율적이지만 트리의 삽입, 삭제 등에는 부담

---

### 순차 검색
<br>

찾는 대상의 값을 처음부터 대조해가면서 맞나 아닌가 체크하는것

하나만 찾는게 아니라면, 처음부터 끝까지 다 해봐야함

대상이 없다면 **검색 실패 = -1 리턴**

````python
# 하나만 찾기

# 함수
import random
def seqSearch(ary, fData) :
    pos = -1
    size = len(ary)
    for i in range(size) :
        if (ary[i] == fData) :
            pos = i
            break
    return pos
    
# 전역변수
dataAry = [ random.randint(1,200) for _ in range(20)]
findData = dataAry[random.randint(0,len(dataAry)-1)]


# 메인코드
print('배열-->', dataAry)
position = seqSearch(dataAry, findData)
if position == -1 :
    print(findData, ' 없어요 ㅠㅠ')
else:
    print(findData, '가', position, '위치에 있음 ^^')
````

### 이진 검색
<br>

정렬이 돼 있어야 함

이진 검색은 전체를 반씩 잘라내서 한쪽을 버리는 방식

시작과 끝이 교차되버리면 검색 실패 = 찾는 값이 안에 없다

````python
# 함수
import random

def binSearch(ary, fData):
    pos = -1
    start = 0
    end = len(ary)-1
    while (start <= end) :
        mid = (start + end) // 2 
        if ary[mid] == fData:
            pos = mid
            break
        elif ary[mid] < fData:
            start = mid + 1
        else:
            end = mid - 1
            
    return pos


# 전역변수
dataAry = [ random.randint(1,200) for _ in range(10)]
dataAry.sort()
findData = dataAry[random.randint(0,len(dataAry)-1)]



# 메인코드
print('배열-->', dataAry)
position = binSearch(dataAry, findData)
if position == -1 :
    print(findData, ' 없어요 ㅠㅠ')
else:
    print(findData, '가', position, '위치에 있음 ^^')

````

# 재귀 호출
<br>

재귀 호출(Recursion)은 자신을 다시 호출하는 것

> 마지막 종이 상자가 나올 때까지 여러 개 겹쳐진 상자를 꺼내 마지막 상자에 반지 넣는 예

> 꺼낸 순서와 반대로 종이 상자를 다시 넣는 예


일반적인 프로그램에서는 무한 반복을 마치고 되돌아가는 조건을 함께 사용함

for 같은 반복문으로도 처리할수 있는데 이왕 개념 배우는 거니까 코드에 신경안써도댐..

````python
# 함수
def openBox():
    global count
    print('상자를 엽니다.')
    count -= 1
    if count == 0 :
        print("*** 선물 넣기 ***")
        return
    openBox()                       # 다시 위로 올라가는 거지만 복사본이 count개만큼 있다고 생각하면 편함.
    print('상자를 닫습니다. $$$')
    return
    
# 전역변수
count = 10


# 메인코드
openBox()
````

````python
# 1에서 10짜리 합계

# 함수
def addNumber(num):
    if num == 1 :
        return 1
    return num + addNumber(num-1)
    
# 전역변수



# 메인코드
print(addNumber(5))


````

````python
# 팩토리얼

# 함수
def factorial(num):
    if num == 1 :
        return 1
    return num * factorial(num -1)
# 전역변수



# 메인코드
print(factorial(5))
````

````python
# 우주선발사

# 함수
def countDown(num):
    if num == 0:
        print('발사')
    else:
        print(num)
        countDown(num-1)
    
# 전역변수



# 메인코드
countDown(5)
````
