예외처리 
```python
try:                                    # 1번만 실행
    예외 발생 가능성 있는 코드
    실행 코드
except 예외타입 클래스 as 객체 :          # 1번 혹은 여러번 실행
    예외처리 코드
else:
    예외가 발생하지 않은 경우 실행 코드
finally:
    예외 유무에 관게 없이 실행되는 코드    # 리소스정리 ex) f.close....
```

1960년대
> 계층형 DBMS : 데이터 참조관계를 계층 구조로 저장 관리(참조 관계를 물리적 포인터, 1:M 참조관계만 가능)

> 망형 DBMS : 데이터 참조관계를 network 구조로 저장 관리(참조 관계를 물리적 포인터, N:M 참조관계 가능)

1970년대
> 수학의 관계 대수를 이용해서 데이터 참조 관계를 모델링 > 데이터 무결성 보장

> RDBMS : oracle

1980년대
> RDBMS 표준 언어 sql 제정

1990년대

> OLTP : transaction 처리 목적, OLAP : Analystic 데이터 마이닝, 의사결정

> Transaction : Unit of work, 분리되어 수행할 수 없는 하나의 작업단위
>> ex) 계좌이체, 온라인 구매
